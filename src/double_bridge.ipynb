# Schwarmintelligenz

* Das zugrunde liegende System besteht aus vielen Individuen.

* Die individuen sind weitgehend homogen, dass heißt sie sind entweder alle identisch oder gehören zu wenigen unterschiedlichen Ausprägungen.

* Die Interaktion zwischen den Individuen basiert auf einfachen Verhaltensregeln, die nur lokale Informationen, welche die Individuen direkt oder über die Umwelt untereinander austauschen. Letzteres wird auch Stigmergie genannt.

* Das Gesamtverhalten des Systems wird durch die Interaktion zwischen den Individuen und ihrer Umwelt erzeugt. Mit anderen Worten, das Gesamtverhalten ist selbstorganisierend.

## Versuchsaufbau

* Für den Versuch wurden mehrere Kolonien, mit je 150-1200 Arbeitern der Argentinischen Ameise *Iridomyrmex humilis*, heran gezogen.

* Ziel der Geometrie ist es, die Pfadauswahl gleich zu gestalten und zu verhindern, dass die Ameise eine Präferenz für einen der beiden Pfade entwickelt.

* Um einen systematischen Fehler vorzubeugen, wurden zwei Module hintereinandergefügt.

# Stochastisches Modell Doppel-Brücken-Experiment

<img style="width:30%;margin:auto" src="double-bridge-extended.png" alt="double-bridge-extended">

Es ergeben sich folgende Iterationsgleichungen für die Pheromonkonzentrationen:

\begin{alignat}{1}
S_j(t) &= S_j(t-1) &\;+\; \Phi\cdot X_{j'}(t-20)            &\;+\; \Phi\cdot X_j(t)\\
L_j(t) &= L_j(t-1) &\;+\; \Phi\cdot\big(1-X_{j'}(t-20r)\big) &\;+\; \Phi\cdot\big(1-X_j(t)\big)
\end{alignat}

| Symbol | Beschreibung |
| ---: | --- |
| $S_j$, $L_j$ | Pheromonkonzentration des kürzeren bzw. längerem Wegs am Wahlpunk $j$ |
| $j$ | Wahlpunkt an der Weggabelung $j\in\{1,2\}$, $j'$ gegenüber liegender Wahlpunkt |
| $\Phi$ | Ameisen pro Sekunde, die die Brücke in beide Richtungen überqueren, hier $0.5s^{-1}$|
| $r$ | Verhältnis (Englisch ratio) von längerem zum kürzeren Weg |
| $X_j(t)$ | 1 falls die Ameise den kürzeren, 0 falls sie den längeren Weg wählt |

# Beispiel

<img style="width:30%;margin:auto" src="double-bridge-extended.png" alt="double-bridge-extended">

Zum Zeitpunkt $t$ wählt eine Ameise den kurzen Weg am Wahlpunkt 1, also $X_{1}(t)=1$.
Bei $t-20$ wählt eine Ameise auch am Wahlpunkt 2 den kurzen Weg, also $X_{2}(t-20)=1$, sodaß sich mit $\Phi=0.5s^{-1}$ ergibt:

\begin{alignat}{1}
S_1(t) &= S_1(t-1) &\;+\; \Phi\cdot 1 + \Phi\cdot 1 = S_1(t-1) + 1\\
L_1(t) &= L_1(t-1) &\;+\; \Phi\cdot 0 + \Phi\cdot 0 = L_1(t-1) + 0
\end{alignat}

# Herleitung aus stetigem Modell

Durch folgende Gleichungen beschreiben Goss, Simon, et al. die Dynamik des stochastischen Systems:

\begin{align}
\frac{dS_j}{dt} &= \Phi\cdot p_{s,j'}(t-20) \;+ \Phi\cdot p_{s,j}(t)\\
\frac{dL_j}{dt} &= \Phi\cdot p_{l,j'}(t-20r) + \Phi\cdot p_{l,j}(t)\\
p_{s,j} &= \frac{(k + S_j)^n}{(k + S_j)^n + (k + L_j)^n}\\
p_{l,j} &= 1 - p_{s,j}
\end{align}

Die Parameter wurden an die Ergebnisse der Doppel-Brücken-Experimente angepasst: $k=20$ und $n=2$.

## Bemerkungen

* $dS_j/dt, dL_j/dt$: Die durchschnittliche Zeit die eine Ameise braucht um die Brücke auf dem kürzeren Weg zu überqueren ist für den benutzten Versuchsaufbau $20$ Sekunden und die Zeidauer für den längeren Weg $20\cdot r$ Sekunden.
* Für die Wahrscheinlichkeiten $p_{s,j}(t)$ und $p_{l,j}(t)$ werden zwei experimental ermittelten Parameter benutzt, zum einem der Dämpfungsparameter $k=20$ (je größer $k$, desto stärker muß die Pheromonspur sein, um die Wegauswahl nicht mehr zufällig wird). Zum anderen der Exponent $n=2$, der den Grad von Nichtliniearität der Wegauswahl bestimmt (ein hoher Wert von $n$ bedeuetet, falls ein Weg nur etwas mehr Pheromonkonzentrat aufweist als der Andere, die nächste Ameise eine hohe Wahrscheinlichkeit hat diesen Weg zu wählen).
* Für $t<0$ sind die Wahrscheinlichkeiten $p_{s,j}(t)$ und $p_{l,j}(t)$ nicht definiert und wir setzten in diesem fall beide Wahrscheinlichkeiten gleich Null.

## Bezeichnungen

* $r$: Verhältnis (Englisch ratio) von längerem zum kürzeren Weg. Z.B. $r=2$, dann ist der längere Weg doppelt so lang wie der Kürzere.

* $j$: Wahlpunkt an der Weggabelung $j\in\{1,2\}$.

* $j'$: Gegenüber liegender Wahlpunk, d.h. $j'=1$ falls $j=2$ und $j'=2$ falls $j=1$.

* $\Phi$: Anzahl Ameisen pro Sekunde, die die Brücke (in beiden Richtingen) überqueren, hier $\Phi=0.5s^{-1}$.

* $S_j$, $L_j$: Pheromonkonzentration des kürzeren bzw. längerem Wegs am Wahlpunk $j$.

* $\frac{dS_j}{dt}$, $\frac{dL_j}{dt}$: Veränderung der Pheromonkonzentration auf dem kürzeren bzw. längerem Weg.

* $p_{s,j}$, $p_{l,j}$: Wahrscheinlichkeit am Wahlpunkt $j$ den kürzeren bzw. längerem Weg auszuwählen.

## Ein diskretes Modell für das Doppelbrückenmodell

Um einen Algorithmus zur Simulation des Doppelbrückenexperiments zu entwerfen, muß das obige stetige Model in ein schrittweises, iteratives Modell überführt werden. Dazu kann man vereinfachend davon ausgehen, dass sich die Zeit in diskreten Zeitpunkten $t=0,1,\ldots,n$ mit jeweils konstanter Zeidauer von $\Delta_t:=1$ Sekunde entwickelt. Da eine Ameise im Durchschnitt mehr als 20 Sekunden braucht, um die Brücke zu überqueren, erscheint diese Annahme gerechtfertigt. Weiter sei $\Delta S_j := S_j(t+1) - S_j(t)=1$ die Änderung der Pheromonkonzentration vom Zeitpunkt $t$ zum nächsten Zeitpunkt $t+1$. Damit lauten dann die beiden Gleichungen für die Pheromonkonzentration (näherungsweise) folgendermaßen:

\begin{alignat}{1}
S_j(t+1) &= S_j(t) + \Phi\cdot p_{s,j'}(t-20) \,&+\, \Phi\cdot p_{s,j}(t)\\
L_j(t+1) &= L_j(t) + \Phi\cdot p_{l,j'}(t-20r)\,&+\, \Phi\cdot p_{l,j}(t)
\end{alignat}

Zur Berechnung der Ameisenanzahl, wird die Zufallsvariable $X_j$ eingeführt:

\begin{align}
X_j(t) &= \begin{cases}
          1 & \text{falls Ameise kürzeren Weg wählt, }P\big(X_j(t)=1\big) = p_{s,j}(t),\\
          0 & \text{falls Ameise längeren Weg wählt, }P\big(X_j(t)=0\big) = 1-p_{s,j}(t).
      \end{cases}
\end{align}

# Python Program zur Simulation des Doppelbrückenmodells

```python
# Importiere benötigte Bibliotheken
import numpy as np
import random
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
```

```python
# Funktion um die Wahrscheinlichkeit für den kurzen Weg zu berechenen
def prob_short(j,t):
    return (20+S[j][t])**2 / ( (20+S[j][t])**2 + (20+L[j][t])**2 )
```

```python
# Funktion um Änderung in Pheromonkonzentration zu berechenen
def pheromon(t,r):
    if t == 0:
        return;

    # Zufallsvariable "würfeln"
    X[0][t] = 0; X[1][t] = 0
    if random.uniform(0, 1) < prob_short(0,t-1):
        X[0][t] = 1
    if random.uniform(0, 1) < prob_short(1,t-1):
        X[1][t] = 1

    # kurzer Weg
    if t - 20 > 0:
        S[0][t] = S[0][t-1] + Phi*X[1][t-20] + Phi*X[0][t]
        S[1][t] = S[1][t-1] + Phi*X[0][t-20] + Phi*X[1][t]
    else:
        S[0][t] = S[0][t-1] + Phi*X[0][t]
        S[1][t] = S[1][t-1] + Phi*X[1][t]
    # langer Weg
    if t - 20*r > 0:
        L[0][t] = L[0][t-1] + Phi*(1-X[1][t-int(20*r)]) + Phi*(1-X[0][t])
        L[1][t] = L[1][t-1] + Phi*(1-X[0][t-int(20*r)]) + Phi*(1-X[1][t])
    else:
        L[0][t] = L[0][t-1] + Phi*(1-X[0][t])
        L[1][t] = L[1][t-1] + Phi*(1-X[1][t])
```

```python
# Durchschnittliche Anzahl Ameisen pro Sekunde (Hin- bzw. Rückweg)
Phi = 0.5
# Anzahl Zeitschritte
t_max = 20*60+1
# Arrays aus zwei Zeilen (zwei Wege) und n Spalten (Anzahl Zeitschritte)
S = np.zeros((2, t_max))
L = np.zeros((2, t_max))
X = np.zeros((2, t_max))
# n Simulation  mit ratio r und Pheromon Konzentration auf langen Weg L_start
def simulation(n=1,r=2.0,p=0, L_start=[0,0]):
    ants_short = np.zeros((n,t_max)) # Anzahl Ameisen auf kurzem Weg
    ants_long = np.zeros((n,t_max))  # Anzahl Ameisen auf langem Weg
    for i in range(0,n):
        S[0][0] = 0; S[1][0] = 0
        L[0][0] = L_start[0]; L[1][0] = L_start[1]
        for t in range(0, t_max):
            pheromon(t=t,r=r)
            if t > 0:
                ants_short[i][t] = Phi * (X[0][t] + X[1][t])
                ants_long[i][t]  = 1 - ants_short[i][t]
    return ants_short, ants_long
```

```python
# Funktion für Graphikausgabe
def plot_histogram(ax, ants_short, r):
    vals = ants_short[:,501:1001].sum(axis=1)
    data = (vals-vals.min())/(vals.max()-vals.min())
    ax.set_ylim(0,1)
    ax.yaxis.set_major_formatter(PercentFormatter(1))
    ax.xaxis.set_major_formatter(PercentFormatter(1))
    ax.tick_params(axis='both', which='major', labelsize=11)
    ax.set_title('$\\Phi={0}, r={1}$'.format(Phi, r), fontsize=16)
    ax.hist(data, 5, weights=np.ones(len(data)) / len(data))
    ax.grid()
```

```python
def plot_single_simulation(s_ants,l_ants, s_phero, l_phero):
    fig, ax = plt.subplots(nrows=1, ncols=1, sharey=True, sharex=True, figsize=(8, 4), dpi=130)
    ax.set_ylim(0,1)
    ax.yaxis.set_major_formatter(PercentFormatter(1))
    ax.plot(s_ants, 'g', label="Anteil Ameisen auf kurzen Weg")
    ax.plot(s_phero, 'g--', label="Pheromonkonzentration auf kurzen Weg")
    ax.plot(l_ants, 'r', label="Anteil Ameisen auf langen Weg")
    ax.plot(l_phero, 'r--', label="Pheromonkonzentration auf langen Weg")
    ax.legend(loc='upper left')
    ax.set_xlabel('Zeit in Sekunden = Anzahl Ameisen ($\\Phi=%g$)' % Phi)
    ax.grid()
```

```python
def create_boxplot_data(ants_short, width=60):
    # Daten für Boxplot erzeugen
    count = 0
    rows = int(ants_short.shape[1]/width)
    cols = ants_short.shape[0]
    data = np.ndarray([rows,cols])
    for t in range(0,int((t_max-1)/width)):
        left = t*width
        right = (t+1)*width
        data[count] = np.mean(ants_short[:,left:right],axis=1)
        #data[0][count] = np.quantile(np.mean(ants_short_20[0:10,1140:1200],axis=1),0.50)
        count += 1
    return data
```

```python
def plot_boxplot(ax, ants_short, r, width=60, outliers=False, left_perc=5, right_perc=95):
    data = create_boxplot_data(ants_short, width=width)
    flierprops=dict(marker='+', markersize=4)
    medianprops = dict(linestyle='-.', linewidth=2.5, color='firebrick')
    # Boxplot zeichnen
    ax.set_ylim(0,1)
    ax.boxplot(list(data[0:data.shape[0]]),whis=[left_perc,right_perc], medianprops=medianprops, flierprops=flierprops)
    ax.set_title('$\\Phi={0}, r={1}$'.format(Phi, r), fontsize=16)
    ax.grid()
    data.shape[0]
```

```python
sims=1000
#np.random.seed(123) # Seed für Pseudozufallszahlen fixieren
(ants_short_10, ants_long_10) = simulation(n=sims, r=1.0, p=1)
(ants_short_14, ants_long_14) = simulation(n=sims, r=1.4, p=1)
(ants_short_20, ants_long_20) = simulation(n=sims, r=2.0, p=1)
(ants_short_20_L, ants_long_20_L) = simulation(n=sims, r=2.0, p=1, L_start=[2000/2,2000/2])
```

```python
(ants_short, ants_long) = simulation(n=1, r=2.0)
s_ants = ants_short[0].cumsum()/(t_max-1) # Anteil Ameisen kurzer Weg
l_ants = ants_long[0].cumsum()/(t_max-1)  # Anteil Ameisen langer Weg
phero_max = S[0].cumsum()[t_max-1] + L[0].cumsum()[t_max-1] # Max. Pheromonkonz.
s_phero = S[0].cumsum()/phero_max         # Anteil Pheromonkonz. kurzer Weg
l_phero = L[0].cumsum()/phero_max         # Anteil Pheromonkonz. langer Weg
```

```python
plot_single_simulation(s_ants,l_ants, s_phero, l_phero)
```
